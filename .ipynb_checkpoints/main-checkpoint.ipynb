```python
# -- IMPORTED --
import pandas as pd
import numpy as np
import csv
import math
import random
#from amplpy import AMPL, DataFrame

# -- FUNCTIONS --


#-- MAIN --
# Create an AMPL instance
#ampl = AMPL()
#config

'''
The format of these data files is:
number of potential warehouse locations (m), number of
customers (n)
for each potential warehouse location i (i=1,...,m): 
capacity, fixed cost
for each customer j (j=1,...,n): demand, cost of allocating
all of the demand of j to warehouse i (i=1,...,m)
'''

# Lectura del archivo
file = "cap41.txt"
settings = []
with open(file) as rf:    
	rf = csv.reader(rf)    
	for row in rf:        
		settings.append(row[0])        
#print("datos",settings)
# Setting principal de cantidad de almacenes y clientes de la primera fila
nWarehouses = int(settings[0].split()[0])
nCustomers = int(settings[0].split()[1])
print(nWarehouses)
print(nCustomers)
```

```python
#TODO: cambiar nombres de variables a nombres correspondientes de la línea 21 y 22/23 de sección anterior
'''
for each potential warehouse location i (i=1,...,m): 
capacity & fixed cost
'''
capacityWarehouse = []
fixedWarehouseCost = []
'''
for each customer j (j=1,...,n): demand & cost of allocating
all of the demand of j to warehouse i (i=1,...,m)
'''
customerDemand = []
costsOfAllocatingDemand = []
# Lectura de primer y segundo valor después de la lectura de la primera fila hasta nWarehouses.
for i in range(1, nWarehouses+1):
    capacityWarehouse.append(int(settings[i].split()[0]))
    fixedCost = settings[i].split()[1]
    fixedCost = int(fixedCost.replace('.', ''))
    fixedWarehouseCost.append(fixedCost)   
print("Capacity of Warehouses:",capacityWarehouse)
print("Fixed Warehouses Cost:", fixedWarehouseCost)

# length: Cantidad de líneas a leer de demandas de i almacenes.
# lines: Cantidad de líneas a leer a partir de la lectura inicial.
length = math.ceil(nWarehouses / 7)
lines = ((1 + length) * nCustomers) + nWarehouses + 1

# Lectura comienza a partir de la linea (nWarehouses + 1) hasta el término del archivo con paso 
# del largo de líneas a leer de demandas de i almacenes del cliente.
'''
Se puede reemplazar o crear una matriz que almacene la demanda del cliente y su costo. 
'''
for i in range(nWarehouses + 1, lines, length + 1):
    numCost = nWarehouses
    customerDemand.append(int(settings[i]))
    #print("Print:", settings[18].split()[1])
    allocatingCostofWarehouseI = []
    count = 0
    for line in range((i + 1), (i + 1 + length)):
        if(numCost >= 7): size = 7
        else: size = numCost
        print("Costs in line", line, ":", end=' ')
        for cost in range(0, size):
            allocatingCostofWarehouseI.append(settings[line].split()[cost])
            print(allocatingCostofWarehouseI[count], end = ' ')
            count += 1
        print('\n')
        numCost -= size
    costsOfAllocatingDemand.append(allocatingCostofWarehouseI)
print('Demanda del cliente \n',pd.DataFrame(customerDemand))
print('Costo de almacenes \n', pd.DataFrame(costsOfAllocatingDemand))
#f = []
#x = []
#d = []
#IcapJ = []

#
#for line in settings:
#    print(line.split()[0])
#    for i in range(1, nWarehouses):
#        capacityWarehouse.append(line.split()[0])
#
```

```python
'''
The objective is to minimize the sum of the opening cost and the assignment cost.
The total demand assigned to a facility must not exceed its capacity
'''

# Random Bit flip generation for initial solution
def bitFlipWarehouse(nWarehouses, fixedWarehouseCost):
    warehouseDisponibility = []
    for i in range(nWarehouses):
        #Si el costo del almacen es 0, significa que debe ser el principal. Sino, cualquier otro
        if(fixedWarehouseCost[i]): warehouseDisponibility.append(1)
        else: warehouseDisponibility.append(random.randint(0,1))
    return warehouseDisponibility

nOfIterations = 1
totalCost = 0
# X = WarehousesAvailable
X = []
    
for n in range(nOfIterations):
    assignmentCost = 0
    X = bitFlipWarehouse(nWarehouses, fixedWarehouseCost)
print(X)
```

```python
def evaluateInitialSolution(X, customerDemand, capacityWarehouse, nWarehouses, nCustomers):
    Y = np.zeros((nWarehouses, nCustomers))
    costs = np.zeros(nWarehouses)
    print(pd.DataFrame(Y))
    for j in range(nCustomers):
        i = random.randint(0,(nWarehouses - 1))
        while((X[i] != 1) or ((costs[i] + customerDemand[j]) > capacityWarehouse[i])):
            i = random.randint(0,(nWarehouses - 1))
        Y[i][j] = 1
        costs[i] += customerDemand[j]
    print('X array: \n', pd.DataFrame(X).T)
    print('Y matrix: \n', pd.DataFrame(Y).T)
    print('Costs \n',pd.DataFrame(costs))
    return 0
openingCost = evaluateInitialSolution(X, customerDemand, capacityWarehouse, nWarehouses, nCustomers)
bestCost = openingCost
```
